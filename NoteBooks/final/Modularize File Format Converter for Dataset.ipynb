```python
import glob
import pandas as pd
from pathlib import Path
import json
import os
```

```python
def get_columns_names(schemas,ds_name,sorting_key='column_position'):
    column_details=schemas[ds_name]
    columns=sorted(column_details,key=lambda col:col[sorting_key])
    return [col['column_name']for col in columns]
```

```python
def read_csv(file,schemas):
    file_path_list=Path(file).parts
    ds_name=file_path_list[-2]
    columns=get_columns_names(schemas,ds_name)
    df=pd.read_csv(file,names=columns)
    return df
```

```python
def convert_to_json(df,tgt_base_dir,ds_name,file_name):
    json_file_path=f'{tgt_base_dir}/{ds_name}/{file_name}'
    os.makedirs(f'{tgt_base_dir}/{ds_name}',exist_ok=True)
    df.to_json(
        json_file_path,
        orient='records',
        lines=True
    )
```

```python
def file_converter(ds_name):
    src_base_dir='data/retail_db'
    tgt_base_dir='data/retail_db_json'
    files=glob.glob(f'{src_base_dir}/{ds_name}/part-*')
    schemas=json.load(open(f'{src_base_dir}/schemas.json'))

    for file in files:
        print(f'Processing {file}')
        df=read_csv(file,schemas)
        file_name=Path(file).parts [-1]
        convert_to_json(df,tgt_base_dir,ds_name,file_name)
```

```python
ds_name='orders'
```

```python
file_converter(ds_name)
```
