```python
import pandas as pd
```

```python
from pathlib import Path
```

```python
src_file_name=glob.glob('data/retail_db/*/part-*')
```

```python
src_file_name
```

```python
schemas = json.load(open('data/retail_db/schemas.json'))
```

```python
base_dir = 'data/retail_db_json'
```

```python
from pathlib import Path
```

```python
for file in src_file_name:
    file_path_list=list(Path(file).parts)
    ds_name=file_path_list[-2]
    file_name=file_path_list[-1]
    json_file_path=f'{base_dir}/{ds_name}/{file_name}'
    print(json_file_path)
```

```python
def get_columns_names(schemas,ds_name,sorting_key='column_position'):
    column_details=schemas[ds_name]
    columns=sorted(column_details,key=lambda col:col[sorting_key])
    return [col['column_name']for col in columns]
    
```

```python
import os
```

```python
tgt_base_dir = 'data/retail_db_json'
for file in src_file_name:
    print(f'Processing {file}')
    file_path_list=list(Path(file).parts)
    ds_name=file_path_list[-2]
    file_name=file_path_list[-1]
    column_names=get_columns_names(schemas,ds_name)
    df=pd.read_csv(file,names=column_names)
    os.makedirs(f'{tgt_base_dir}/{ds_name}',exist_ok=True)
    json_file_path=f'{tgt_base_dir}/{ds_name}/{file_name}'
    df.to_json(
        json_file_path,
        orient='records',
        lines=True
    )
```
